```python
from xvla_wlr.model_legacy import XVLA, XVLAProcessor

model = XVLA.from_pretrained("2toINF/X-VLA-SoftFold")
processor = XVLAProcessor.from_pretrained("2toINF/X-VLA-SoftFold", use_fast=True)
```

```python
import torch
# TODO add option for bfloat16 in pipeline
# model = model.to(device="cuda", dtype=torch.bfloat16)
model = model.to(device="cuda")
```

```python
import torch
from xvla_wlr.model_legacy import compute_actions

compute_actions_ = torch.compile(
    compute_actions,
    mode="max-autotune",
    disable=True,
)
```

```python
from xvla_wlr.model_legacy import Observation

compute_actions_(
    model=model,
    processor=processor,
    observation=Observation.sample(),
)
```
